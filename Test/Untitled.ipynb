```python
s = "india is great country".
```

```python
def english_tokenizer(text):
    return [x.text for x in nlp.tokenizer(text)]
```

```python
field =  torchtext.data.Field(tokenize = english_tokenizer ,
                              preprocessing =lambda x :["start"]+x+["end"])

a = torchtext.data.Example.fromlist([s] , [("text",field)])
```

```python
TEXT = torchtext.data.Field(tokenize=lambda x :x.split(),include_lengths=True,
                           batch_first=True)
CLass_LABEL = torchtext.data.LabelField(use_vocab=False)
REGr_LABEL = torchtext.data.LabelField(use_vocab=False,dtype=torch.float32)

dataset = torchtext.data.TabularDataset("Test/te.csv" , format="csv",
                              fields={"text":("TEXT",TEXT) ,
                                      "joy_class":("CLASS",CLass_LABEL),
                                         "joy_reg":("REGR",REGr_LABEL)}
                                       )
TEXT.build_vocab(dataset)
# single field represents a particular set of data. 10 examples across 1 Field
# 10 examples will be used to build the vocab
```

```python
it = torchtext.data.BucketIterator(dataset,batch_size=8,shuffle=True,)
```

```python
c = next(iter(it))
c
```
